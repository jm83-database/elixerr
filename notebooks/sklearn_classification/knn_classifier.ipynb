```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.neighbors import KNeighborsClassifier
```

```python
# 기존 데이터 닥스훈트와 진돗개
dach_length = [55, 57, 64, 63, 58, 49, 54, 61]
dach_height = [30, 31, 36, 30, 33, 25, 37, 34]
 
jin_length = [56, 47, 56, 46, 49, 53, 52, 48]
jin_height = [52, 52, 50, 53, 50, 53, 49, 54]

# 새로운 데이터
newdata_length = [59]
newdata_height = [35]
```

```python
plt.scatter(dach_length, dach_height, label='Dachshund')
plt.scatter(jin_length, jin_height, label='jindoDog')
plt.scatter(newdata_length, newdata_height, label='new', marker='s')
plt.legend()
plt.show()
```

```python
# 데이터 테이블로 만들기
d_data = np.column_stack((dach_length,dach_height))
d_label = np.zeros(len(d_data))

j_data = np.column_stack((jin_length,jin_height))
j_label = np.ones(len(j_data))

dogs = np.concatenate((d_data, j_data))
labels = np.concatenate((d_label,j_label))
dogs, labels
```

```python
k = 3
knn = KNeighborsClassifier(n_neighbors=k)
knn
```

```python
knn.fit(dogs,labels)
```

```python
newdata = np.column_stack((newdata_length,newdata_height))
y_pred = knn.predict(newdata)
y_pred
```

```python
dog_classes = {0:'닥스훈트',1:'진도개'}
print(f'예측데이터 : {newdata}, 예측결1과 : {dog_classes[y_pred[0]]}')
```

```python
# 진도개 예측 테스트
newdata1 = np.array([[50,41]])
newdata1
```

```python
for k in [1, 3, 5, 9]:
    knn = KNeighborsClassifier(n_neighbors=k)
    knn.fit(dogs,labels)
    y_pred1 = knn.predict(newdata1)
    print(f'k={k} 일때 예측데이터 : {newdata1}, 예측결1과 : {dog_classes[y_pred1[0]]}')
```
