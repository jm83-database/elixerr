# 데이터 불러오기

```python
import requests

base_url = 'https://search.daum.net/search?nil_suggest=btn&w=tot&DA=SBC&q=' # 기본 주소

location = '후암동' # 검색하고 싶은 지역
search_query = location + '+날씨'

# location = input('지역 입력 예) 한남동 : ')
# base_url = f'https://search.daum.net/search?nil_suggest=btn&w=tot&DA=SBC&q={location}+날씨'


url = base_url + search_query
```

```python
url # 합친 결과 확인
```

```python
html_weather = requests.get(url)
html_weather
```

```python
html_weather.text # 내용 확인
```

```python
# beatifulsoup으로 파싱
soup_weather = BeautifulSoup(html_weather.text, 'lxml')
soup_weather
```

# 날씨 정보 추출

```python
# 날씨 정보 추출(온도)
txt_temp = soup_weather.select_one('strong.txt_temp').text
txt_temp
```

```python
# 기상상태 추출
txt_weather = soup_weather.select_one('span.txt_weather').text
txt_weather
```

```python
# 풍속, 습도, 미세먼지정보 추출
dl_weather = soup_weather.select('dl.dl_weather>dd')
dl_weather
```

```python
# 여러줄 코드
dl_list = []
for value in dl_weather:
    print(value.text)
    dl_list.append(value.text)

dl_list
```

```python
# 한줄 코딩
[wind_speed, humidity, pm10] = [x.text for x in dl_weather]
print(f'''현재 풍속: {wind_speed}
현재 습도 : {humidity}
미세먼지 : {pm10}
''')
```
