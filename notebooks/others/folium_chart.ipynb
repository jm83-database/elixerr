```python
import folium

map = folium.Map(location=[37.4965, 126.9573], 
                zoom_start=17,
                width=750,
                height=500)


folium.Marker(location=[37.4965, 126.9573],
            popup='숭실대학교',
            tooltip='숭실대학교',
            icon=folium.Icon(color='red',icon='horse',prefix='fa')).add_to(map)
map.save('map.html')
```

```python
import pandas as pd

df = pd.read_excel('../../data/서울시_반려견_놀이터.xlsx')
df
```

```python
# 서울시 반려견 놀이터 위치
lat = df['Latitude'].mean()
lon = df['Longitude'].mean()
dog_park_map = folium.Map(location=[lat, lon],
                            zoom_start=11,
                            width=750,
                            height=500)


for n, row in df.iterrows():
    folium.Marker(location=[row['Latitude'], row['Longitude']],
            popup=row['공원명'],
            tooltip=row['공원명'],
            icon=folium.Icon(color='red',icon='star',prefix='fa')).add_to(dog_park_map)

dog_park_map.save('dog_park_map.html')
```
