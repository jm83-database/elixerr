```python
import pandas as pd
import matplotlib.pyplot as plt
plt.rc('font', family='Malgun Gothic')
```

```python
data = pd.read_excel('../../data/fine_dust.xlsx', index_col='area')
data
```

```python
# 2018년 미세먼지 그래프
data2018 = data[2018]
plt.figure(figsize=(15,4))
plt.plot(data2018,color='b',marker='o')
plt.xlabel('area')
plt.ylabel('micrometer')
plt.title('2018 fine dust line graph')
plt.grid()
plt.show()
```

```python
# 2016~2019년 미세먼지 그래프
plt.figure(figsize=(15,4))

for year in range(2016,2020):
    chardata=data[year]
    plt.plot(chardata,marker='o',label=year)

plt.xlabel('area')
plt.ylabel('micrometer')
plt.title('2016-2019 fine dust line graph')
plt.legend()
plt.grid()
plt.show()
```

```python
# 2017 막대그래프
data2017 = data[2017]
plt.figure(figsize=(15,4))
plt.bar(x=data2017.index,height=data2017,color='g')
plt.ylim(35,55) # y축 범위
plt.xlabel('area')
plt.ylabel('micrometer')
plt.title('2017 fine dust bar graph')
plt.grid()
plt.show()
```

```python
# 산점도 그래프

data2 = pd.read_excel('../../data/health_screenings_2020_1000ea.xlsx')
data2
```

```python
data2.info()
```

```python
heightdata = data2['height']
weightdata = data2['weight']

plt.figure(figsize=(15,4))
plt.scatter(x=heightdata,y=weightdata)
plt.show()
```

```python
data2.columns
# systolic : 수축기 혈압
# diastoloc : 이완시 혈압
# blood_sugar : 공복 시 혈당
```

```python
# 수축기 혈압, 이완기 혈압, 공복 혈당 그룹 산점도 그래프

systolic_data = data2['systolic']
diastolic_data = data2['diastolic']
blood_sugar_data = data2['blood_sugar']

plt.figure(figsize=(10,6))
plt.scatter(x=systolic_data,y=diastolic_data,color='r',label='systolic*diastolic')
plt.scatter(x=systolic_data,y=blood_sugar_data,color='g',label='systolic*blood_sugar')
plt.scatter(x=diastolic_data,y=blood_sugar_data,color='b',label='diastolic*blood_sugar')
plt.xlim(40,180)
plt.ylim(30,300)
plt.title('2020 health screenings group scatter graph')
plt.legend()
plt.grid()
plt.show()
```

```python
# 남성의 성별, 키 데이터로 남성 키 히스토그램
# 남성 데이터셋 생성
male_data = data2.loc[data2['gender']==1,['gender','height']]

plt.figure(figsize=(10,6))
plt.hist(male_data['height'],bins=8,color='y',label='male') # bins : 막대 구간, 
plt.title('2020 Health screenings male height histogram')
plt.legend()
plt.grid()
plt.show()
```

```python
# 남성 여성의 키 데이터로 히스토그램
# 여성 데이터셋 생성
female_data = data2.loc[data2['gender']==2,['gender','height']]

plt.figure(figsize=(10,6))
plt.hist(male_data['height'],bins=8,color='y',label='male') # bins : 막대 구간
plt.hist(female_data['height'],bins=8,color='b',label='female',alpha=0.3) # bins : 막대 구간, alpha : 투명도
plt.title('2020 Health screenings gender height histogram')
plt.legend()
plt.xlabel('height')
plt.ylabel('frequency')
plt.grid()
plt.show()
```

```python
# 상자수염 그래프 작성(boxplot)
import numpy as np

height = np.array([male_data['height'],female_data['height']],dtype=object)

plt.figure(figsize=(10,6))
plt.boxplot(height,tick_labels=['male','female'])
plt.title('2020 Health screenings gender height boxplot')
plt.ylim(130,200)
plt.grid()
plt.show()
# 위 아래 수염을 벗어난 것은 이상치, 수염의 끝은 최대값, 최소값, 중앙애 선은 중앙값, 상자의 위쪽은 75%, 아래는 25% 지점
```
