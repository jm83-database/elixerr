```python
import bikes_data

bikes = bikes_data.data_preprocessing()
```

# 인기 있는 대여소 분석
- 주말에 인기 있는 대여소 top 50
- 평일에 인기 있는 대여소 top 50

```python
bikes.columns
```

```python
rent_bike = bikes.pivot_table(index=['대여 대여소명','대여점위도','대여점경도'],columns=['주말구분'],values='자전거번호',aggfunc='count')
rent_bike.head()
```

```python
rent_bike.info()
```

```python
rent_bike.sort_values('주말',ascending=False).head(3)
```

```python
weekend_favorites50 = rent_bike.nlargest(50,'주말')[['주말','평일']].reset_index()
weekend_favorites50
```

```python
weekend_favorites50.columns
```

```python
import folium

map = folium.Map( location=[bikes['대여점위도'].mean(), bikes['대여점경도'].mean()], zoom_start=11, width=700, height=700)
for i in weekend_favorites50.index:
    sub_lat = weekend_favorites50.loc[i,'대여점위도']
    sub_lon = weekend_favorites50.loc[i,'대여점경도']
    name = weekend_favorites50.loc[i,'대여 대여소명']
    folium.Marker(
        location=[sub_lat, sub_lon],
        popup=name,
    ).add_to(map)
map
```

# 평일 인기있는 대여소 시각화

```python
days_favorites50 = rent_bike.nlargest(50,'평일')[['주말','평일']].reset_index()
```

```python
for i, row in days_favorites50.iterrows():
    sub_lat = row['대여점위도']
    sub_lon = row['대여점경도']
    name = row['대여 대여소명']
    folium.Marker(
        location=[sub_lat, sub_lon],
        popup=name, 
        icon=folium.Icon(color = 'red', icon = 'bicycle')
    ).add_to(map)
    
map
```
