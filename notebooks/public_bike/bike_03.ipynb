```python
import bikes_data

bikes = bikes_data.data_preprocessing()
```

# 1. 시간대 & 대여구 데이터분석

```python
bikes.columns
```

```python
# 시간적 & 장소적 특징 분석
hourly_gu_use = bikes.pivot_table(index='대여시간대', columns='대여구', values='자전거번호', aggfunc='count')
hourly_gu_use.head()
```

```python
hourly_gu_use.plot()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=(10,8))
sns.heatmap(hourly_gu_use.T)
```

```python
plt.figure(figsize=(15,8))
sns.boxplot(hourly_gu_use)
```

# 2. 일자 & 대여구 데이터분석

```python
daily_gu_use = bikes.pivot_table(index='일자', columns='대여구', values='자전거번호', aggfunc='count')
daily_gu_use.head()
```

```python
plt.figure(figsize=(10,8))
sns.heatmap(daily_gu_use.T)
plt.figure(figsize=(15,8))
sns.boxplot(daily_gu_use)
```
